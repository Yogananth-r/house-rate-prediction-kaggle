```python
import matplotlib as mlp
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from matplotlib import rcParams
rcParams['figure.figsize'] = 10, 5
```

```python
columns = xtrain.select_dtypes(include=['int']).columns.tolist()
print(columns)

for column in columns[:]:
    if xtrain[column].value_counts().shape[0] > 20:
        columns.remove(column)
        
for column in columns:
    sns.countplot(x=column, data=xtrain)
    
    plt.title(column)
    plt.show()
```

```python
sns.histplot(x='SalePrice', kde=True, data=xtrain)
plt.title('Sale Price')
plt.show()
```

```python
x = xtrain.drop(columns=['SalePrice'])
y = xtrain.SalePrice
print('x Shape:', x.shape)
print('y Shape:', y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
print('x train Shape:', x_train.shape)
print('y train Shape:', y_train.shape)
print('x test Shape:', x_test.shape)
print('y test Shape:', y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=100)
rfr.fit(x_train, y_train)
print('Score:', rfr.score(x_test, y_test))
yhat_rfr = rfr.predict(x_test)
```

```python
ax1 = sns.distplot(xtrain['SalePrice'], hist=False, color='r', label='Actual Price')
sns.distplot(yhat_rfr, hist=False, color='b', label='Fitted Price', ax=ax1)

plt.title('Actual price vs Fitted price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Features')
plt.show()
```

```python
col1 = xtest.Id
col2 = rfr.predict(xtest)
df1 = pd.DataFrame(col1, columns=['Id'])
df2 = pd.DataFrame(col2, columns=['SalePrice'])

results = pd.concat([df1, df2], axis=1, join='inner')
results.to_csv('Submission.csv', index=False)
```
